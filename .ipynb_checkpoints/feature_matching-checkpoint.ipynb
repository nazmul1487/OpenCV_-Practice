```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('images/opencv-feature-matching-template.jpg',0)
img2 = cv2.imread('images/opencv-feature-matching-image.jpg',0)
```

```python
orb = cv2.ORB_create()   #ord is detector

#finding keypoints and descriptors for those pictures
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)

#bf BFMatcher objects which use for matching
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None, flags=2)
plt.imshow(img3)
plt.show()

# cv2.imshow('img3',img3)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
